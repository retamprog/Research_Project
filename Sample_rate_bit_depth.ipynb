```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# signal parameters
freq = 5
duration = 1
analog_res = 1000

t_analog = np.linspace(0,duration,analog_res)
# print(t_analog)
'''
creating the sine wave using the mathematical sine wave formulae
y(t) = Asin(wt + phi)
y(t) = Asin(2*pi*f*t + phi)

where:
y -- > the output function or the y plot of the sine wave
w -- > called as omega or the angular frequency of wave
w = 2*pi*f  --> (the rate of change in radians per second or the number of revolutions per second)
f=1/T (T - the time period of one revolution)
phi = phase , the intial angle in radians

'''
y_analog = np.sin(2*np.pi*freq*t_analog)

plt.figure(figsize=(12,5))
plt.plot(t_analog,y_analog,label='Analog signal',linewidth=2)
plt.grid()
plt.show()
```

```python
sr_low =7
sr_high = 80

print(1/sr_low)
t_low = np.arange(0,duration,1/sr_low)
t_high = np.arange(0,duration,1/sr_high)
y_low = np.sin(2*np.pi*freq*t_low)
y_high = np.sin(2*np.pi*freq*t_high)

plt.figure(figsize=(12,5))
plt.plot(t_low,y_low,label='low sample rate diagram',linewidth=2)
plt.show()
plt.plot(t_high,y_high,label = 'High sample rate diagram',linewidth=2)
plt.grid()
plt.show()
```
